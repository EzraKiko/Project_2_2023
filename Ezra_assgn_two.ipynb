```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd. read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

1. Show a breakdown of distance from home by job role and attrition.

```python
dataset.groupby(['Attrition','JobRole'])['DistanceFromHome'].sum()
```

2. Compare average monthly income by education and attrition.

```python

dataset.groupby(['Attrition','Education'])['MonthlyIncome'].mean()
```

#### ASIGNMENT THREE
> Based on your analysis in Assignment 2, express your results in the form of the most appropriate visualization using any visualization library of your choice.

```python
# Visualisation for question 1
sns.heatmap(dataset.groupby(['Attrition','JobRole'])['DistanceFromHome'].sum().unstack(),annot=True,fmt=".0f")
```

```python
# Visualisation for question 2
dataset.groupby(['Attrition','Education'])['MonthlyIncome'].mean().plot(kind='bar')
plt.xlabel('Attrition&JobRole')
plt.ylabel('frequency')
plt.show()
```
